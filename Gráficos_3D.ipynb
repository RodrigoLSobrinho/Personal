```python
df = pd.read_csv("Outputs_Planilhas/1-ALS-1-AL.csv")
df
```

#### Gráficos 3D 

```python
%matplotlib
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df["IH"],df["Tmax"],df["COT"], c="black", s=15)
ax.set_xlim(0,1000)
ax.set_ylim(350,550)
ax.set_zlim(0.1,20)
ax.set_xlabel("IH")
ax.set_ylabel("Tmax")
ax.set_zlabel("COT (w%)")
```

```python
%matplotlib
df["S2/S3"]=df["S2"]/df["S3"]
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df["IH"],df["S2/S3"],df["Profundidade"], c="black", s=15)
ax.set_xlim(10,1000)
ax.set_ylim(0.1,100)
#ax.set_xscale("log")
#ax.set_yscale("log")
ax.set_zlim(ax.get_zlim()[::-1])
ax.set_xlim(ax.get_xlim()[::-1])
ax.set_xlabel("IH")
ax.set_ylabel("S2/S3")
ax.set_zlabel("Profundidade (m)")
```
